# Match case 

```python
# Match is used to replace multiple if conditions.
# Match is more powerful then switch case in C & C++
# match can work if conditions and data structure.
```

```python
# Example-1: Structure of match
num = int(input("Enter a number.."))

match num:
    case 10:
        print("Vanilla")
    case 10:
        print("Chocolate")
    case 30:
        print("Strawberray")
    case _:
        print("Anything will work")
```

```python
# Example-2: using Data structure in match
data = [10, 20, 30]

match data:
    case [30, 20, 10]:
        print("Hello")
    case [10, 20, 30]:
        print("Hi")
    case _:
        print("Hello Hi")
```

```python
# Example-3: conditions in match

num = int(input("Enter a Number.."))

match num:
    case _ if num > 0:
        print(f"Its a Positive Number: {num}")
    case _ if num < 0:
        print(f"Its a Negative Number: {num}")
    case _:
        print("Its a Zero")
```

```python
num = int(input("Enter a Number.."))

match num:
    case x if x > 0:
        print(f"Its a Positive Number: {num}")
    case y if y < 0:
        print(f"Its a Negative Number: {num}")
    case _:
        print("Its a Zero")
```

```python
boy_expectation = {"Intelligent", "Smart", "Funny", "Only_daughtor", "Rich"}
girl_expectation = {"Smart", "Rich", "Funny", "Loyal"}
boy_reality = {"Bankrupt", "Funny", "Loyal", "Long_beard"}
instagram_profile = {"Smart", "Rich", "Loyal", "Funny"}
aadhar_card = {"Funny", "Loyal"}

# equivalent operation
if(girl_expectation == boy_expectation):
    print("Rishta Pakka")
else:
    print("Who are you")

if(girl_expectation == instagram_profile):
    print("Congratulations")
else:
    print("Who are you")

# not equivalent operation
if(instagram_profile != aadhar_card):
    print("You are Big Lier")
else:
    print("You are great person")
```

```python
# Python follow the LEGB rule for scoping

# L : Python first look the variable into local scope
# E : Enclosed scope : within the current function or parent function or grand parent or Mutata
# G : Global Scope (Module level)
# B : Built in Scope (check inbuilt functions e.g len(), sum())
```

# lambda functions

```python
# Def: A lambda function is a single line function.
# used to perform some specific small operation.
# lambda function popularly used with other functions.
# lambda function don't have any name
```

```python
# Example-1
obj = lambda a : a + 100

print(obj(10))
```

```python
# Example-2
obj2 = lambda a,b,c : a + b + c

print(obj2(100, 200, 300))
```

```python
# Example-3 : use lambda function inside the other function
def fun(power):
    return lambda a : a**power

obj1 = fun(2)
obj2 = fun(3)
obj3 = fun(4)

print(obj1(10))
print(obj2(10))
print(obj3(10))
```

```python
# Example-4: if-else

val = lambda a, b : a if a > b else b

print(val(10, 20))
```

```python
# Example-5 : lambda function for sorting list of tuples

students = [('Narshima', 1000), ('Ram', 100), ('Amruth', 200), ('Bala', 30000)]

sorted_students = sorted(students, key = lambda x : x[1])

print(sorted_students)
```

```python
# Example-6 : sort dictionary with lambda

Gf_dict = {'Narsima' : 6, 'Amruth' : 7, 'Ram' : 5, 'Naveen': 200}

sorted_dict = dict(sorted(Gf_dict.items(), key = lambda x : x[1]))

print(sorted_dict)
```

### Lambda function with filter function

```python
# filter function: filter function, Filter the list of elements based upon a condition
# filter function get applied to all the elements of input list

# Synatx of filter function: filter(condition, data)

wake_up = [7, 8, 6, 5, 9, 3, 2, 4]

early_birds = list(filter(lambda a : a <= 6, wake_up))

print(early_birds)
```

```python
wake_up = [7, 8, 6, 5, 9, 3, 2, 4]

early_birds = list(filter(lambda a : a%2 == 0, wake_up))

print(early_birds)
```

```python
# nested filter function
wake_up = [7, 8, 6, 5, 9, 3, 2, 4]

early_birds = list(filter(lambda a : a%2 == 0, filter(lambda a : a <= 6, wake_up)))

print(early_birds)
```

```python
wake_up = [7, 8, 6, 5, 9, 3, 2, 4]

early_birds = list(filter(lambda a : a%2 == 0, filter(lambda a : a <= 6, filter(lambda a : a < 9, wake_up))))

print(early_birds)
```

### lambda with map() function

```python
# map function: map transform all the input elements based upon the expression 

shorts_watch = [100, 2, 10, 20, 999, 1000]

target = list(map(lambda a : a//2, shorts_watch))

print(target)
```

```python
# nested map

shorts_watch = [100, 2, 10, 20, 999, 1000]

target = list(map(lambda a : a*3, map(lambda b : b//2,shorts_watch)))

print(target)
```

```python
shorts_watch = [100, 2, 10, 20, 999, 1000]

target = list(map(lambda a : a//2, filter(lambda b : b > 20,shorts_watch)))

print(target)
```

### lambda with reduce()

```python
# Def: reduce function works cumulativaley to find singular result

from functools import reduce
```

```python
Hours_study = [2, 3, 3, 1, 2, 1, 60, 50, 10, 12, 24, 5]

val = reduce(lambda a, b : a / b, Hours_study)

print(val)
```

```python
Hours_study = [2, 3, 3, 1, 2, 1, 60, 50, 10, 12, 24, 5]

val = reduce(lambda a, b : a + b, map(lambda a : a//2, filter(lambda b : b > 20,Hours_study)))

print(val)
```
