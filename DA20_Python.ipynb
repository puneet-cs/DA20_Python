# Match case 

```python
# Match is used to replace multiple if conditions.
# Match is more powerful then switch case in C & C++
# match can work if conditions and data structure.
```

```python
# Example-1: Structure of match
num = int(input("Enter a number.."))

match num:
    case 10:
        print("Vanilla")
    case 10:
        print("Chocolate")
    case 30:
        print("Strawberray")
    case _:
        print("Anything will work")
```

```python
# Example-2: using Data structure in match
data = [10, 20, 30]

match data:
    case [30, 20, 10]:
        print("Hello")
    case [10, 20, 30]:
        print("Hi")
    case _:
        print("Hello Hi")
```

```python
# Example-3: conditions in match

num = int(input("Enter a Number.."))

match num:
    case _ if num > 0:
        print(f"Its a Positive Number: {num}")
    case _ if num < 0:
        print(f"Its a Negative Number: {num}")
    case _:
        print("Its a Zero")
```

```python
num = int(input("Enter a Number.."))

match num:
    case x if x > 0:
        print(f"Its a Positive Number: {num}")
    case y if y < 0:
        print(f"Its a Negative Number: {num}")
    case _:
        print("Its a Zero")
```

# 2. Comparison (== , !=, <=, <, >=, >)

```python
boy_expectation = {"Intelligent", "Smart", "Funny", "Only_daughtor", "Rich"}
girl_expectation = {"Smart", "Rich", "Funny", "Loyal"}
boy_reality = {"Bankrupt", "Funny", "Loyal", "Long_beard"}
instagram_profile = {"Smart", "Rich", "Loyal", "Funny"}
aadhar_card = {"Funny", "Loyal"}

# equivalent operation
if(girl_expectation == boy_expectation):
    print("Rishta Pakka")
else:
    print("Who are you")

if(girl_expectation == instagram_profile):
    print("Congratulations")
else:
    print("Who are you")

# not equivalent operation
if(instagram_profile != aadhar_card):
    print("You are Big Lier")
else:
    print("You are great person")
```

```python
boy_expectation = {"Intelligent", "Smart", "Funny", "Only_daughtor", "Rich"}
girl_expectation = {"Smart", "Rich", "Funny", "Loyal"}
boy_reality = {"Bankrupt", "Funny", "Loyal", "Long_beard"}
instagram_profile = {"Smart", "Rich", "Loyal", "Funny"}
aadhar_card = {"Funny", "Loyal"}

# subset operation
if(girl_expectation <= instagram_profile):
    print("It is Subset")
else:
    print("Not a subset")

# proper subset
if(aadhar_card < girl_expectation):
    print("It is Proper subset")
else:
    print("Not proper subset")
```

# 3. Union

```python
# Def: take all the elements of all sets, 
# but common element will be consider one time.
```

```python
set1 = {1,2,3,4,5}
set2 = {3,4,5,6,7}
set3 = {'a', 'b', 'c', 1,2,3,4}

set4 = set1 | set2

set5 = set1.union(set2)

set6 = set1 | set2 | set3

# update() : Inplace union operation
# inplace : which don't take extra space

set1.update(set2)
print(set1)

set2 = {3,4,5,6,7}
set3 = {'a', 'b', 'c', 1,2,3,4}

set1.update(set2,set3)
print(set1)
```

```python
# using |= operator for inplace union operation
set2 |= set3   # set2.update(set3)  =>  set2 = set2 | set3
print(set2)
```

```python
# Interview Question: What is the difference between | and union()
set1 = {1,2,3,4,5}
tuple2 = (3,4,5,6,7)
list3 = ['a', 'b', 'c', 1,2,3,4]
```

```python
# union() function can work with different data structures
# But union operator only works with sets
set5 = set1.union(tuple2,list3)
print(set5)
```

```python
set4 = set1 | tuple2 | list3
print(set4)
```
