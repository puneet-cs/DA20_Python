# Type casting

```python
# integer type casting
v1 = int(10.5)  # float to int
v2 = int("30")  # str to int
v3 = int(True)  # Bool to int
v4 = int(False)  # Bool to int
v5 = int("101", base = 2)
v6 = int("F", base = 16)

print(v1, type(v1))
print(v2, type(v2))
print(v3, type(v3))
print(v4, type(v4))
print(v5, type(v5))
print(v6, type(v6))
```

```python
# float type casting
v7 = float(1000)
v8 = float("40")
v9 = float(True)

print(v7, type(v7))
print(v8, type(v8))
print(v9, type(v9))
```

```python
# string type casting
v10 = str(20)
v11 = str(10.55)
v12 = str(True)

print(v10, type(v10))
print(v11, type(v11))
print(v12, type(v12))
```

```python
# Bool type cast
def fun():
    return "Hello"

v13 = bool(10)
v14 = bool(10.55)
v15 = bool("Hello")
v16 = bool(0)
v17 = bool("")
v18 = bool(fun())

print(v13, type(v13))
print(v14, type(v14))
print(v15, type(v15))
print(v16, type(v16))
print(v17, type(v17))
print(v18, type(v18))
```

# Assignment operators

```python
# += , -=, *=, /=, %=, //+, &=, |=, ^=, >>=, <<=, :=
```

```python
a = 10

a += 5   # a = a + 5
print(a)

a -= 3   # a = a - 3
print(a)

a *= 3   # a = a * 3
print(a)

a /= 3   # a = a / 3
print(a)

a %= 5   # a = a % 5
print(a)

a = 24

a //= 5   # a = a // 5
print(a)

a &= 5   # a = a & 5
print(a)

a |= 5   # a = a | 5
print(a)

a ^= 5   # a = a ^ 5
print(a)

a = 10

a <<= 4   # a = a << 4
print(a)

a >>= 4   # a = a >> 4
print(a)

print(x := 20)
```

```python
# Basic interview Question
if(n := 10) > 5:
    print("Hello")
else:
    print("Hi")
```

# Identity operators

```python
# is, is not

a = [1,2,3]
b = [1,2,3]
c = a

print(a == b)
print(a is b)
print(a is c)

print(a is not b)
print(a is not c)
print(b is not c)
```

# Membership operator

```python
# in, not in

fruits = ['Simaregipondu', 'Aratipandu', 'danimma']

print('danimma' in fruits)
print('Banana' not in fruits)
```
